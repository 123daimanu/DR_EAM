```python
from matplotlib import pyplot as plt
from matplotlib import rcParams
import numpy as num
import pandas as pan
%matplotlib inline
```

```python
fig1=plt.figure(figsize=(4,4))
ax1=fig1.add_subplot(2,1,1)
ax2=fig1.add_subplot(2,1,2)

#---------------------------------------
#subplot adjust
left  =0.1# the left side of the subplots of the figure
right = 0.9    # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.2  # the amount of width reserved for blank space between subplots,
               # expressed as a fraction of the average axis width
hspace = 0.2  # the amount of height reserved for white space between subplots,
               # expressed as a fraction of the average axis height
    
fig1.subplots_adjust(left=left,right=right,bottom=bottom,top=top,wspace=wspace,hspace=hspace)
plt.show()
```

```python
#subplot2grid
x=num.linspace(0,1,100)
fig2=plt.figure(figsize=(8,8))
ax=fig2.add_axes([0,0,1,1])
ax3=plt.subplot2grid((3,3),(0,0),colspan=1,rowspan=1)
ax4=plt.subplot2grid((2,3),(0,1),rowspan=1,colspan=1)
ax5 = plt.subplot2grid((3, 3), (1, 0),rowspan=1,colspan=1)


ax3.plot(x,x**2)
ax4.plot(x)
ax5.plot(num.log(x))
```

```python
#two different plots in same graph with different units of y

fig=plt.figure(figsize=(5,5))
ax1=fig.add_subplot(1,1,1)
ax2=ax1.twinx()
ax1.plot(x,x**2,'r-',linewidth=2)
ax2.plot(x,num.exp(x),'b-',linewidth=2)
ax1.set_ylabel("density",color="red",fontsize=12)
ax2.set_ylabel("temperature",color="blue",fontsize=12)
ax1.set_xlabel("time",fontsize=12)
fig.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import rcParams
import numpy as num

%matplotlib inline
#setting different parameteres
params = {
    'font.family':'serif',
   'axes.labelsize': 12,
   'font.size': 12,
   'legend.fontsize': 12,
   'xtick.labelsize': 12,
   'ytick.labelsize': 12,
   'text.usetex': False,
   'figure.figsize': [4.5, 4.5]
   }
rcParams.update(params)


#plotting one inside other
x1=num.linspace(0,10,1000)
fig1=plt.figure()

ax1=fig1.add_axes([0,0,1,1])
ax2=fig1.add_axes([1.2,.65,.3,.3])
y1=num.sin(x1)/x1
ax1.plot(x1,y1,linewidth=2,color="red")
ax1.axhline(y=0,color="black",linewidth=1.5,linestyle="dashed")
#ax1.fill_between(x,.25,alpha=.4)
ax1.axvspan(5,7,alpha=0.1)
#ax1.arrow(6.3,0,-2,.2,shape='full', lw=1, length_includes_head=True, head_width=0)
#ax1.text(1, 0.9,'Zeros of\n ${\sin(x)}/{x}$', horizontalalignment\
#    ='center', verticalalignment='center', transform=ax.transAxes)
ax1.annotate("Zeros",xy=(6.3,0),xytext=(4,.1),arrowprops=dict(facecolor='black',shrink=0.01,headwidth=5,width=1),)
ax1.set_xlabel("x",fontsize=12)
ax1.set_ylabel(r"$\alpha$",fontsize=12)
ax1.set_frame_on(True)
ax1.invert_xaxis()
ax1.invert_yaxis()



x2=num.linspace(5,7,100)
ax2.plot(x2,num.sin(x2)/x2)
ax2.axhline(y=0,color="black",linewidth=1.5,linestyle="dashed")
ax2.set_xticks([5,6,7])
ax2.set_yticks([-0.1,0,.1])
ax2.axvline(x=6.3,color="black",linestyle="dashed",linewidth="1.5")
ax2.axvspan(5,7,alpha=0.1)
plt.show()
```

```python
#Adding color bar
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.6,0.8])
image=num.random.poisson(10,(100,80))
i=ax.imshow(image,interpolation="nearest",aspect="auto")
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_xticks([])
ax.set_yticks([])
colorbar_ax=fig.add_axes([0.7,0.1,0.05,0.8])
fig.colorbar(i,cax=colorbar_ax)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')


data = [
    go.Surface(
        z=z_data.as_matrix()
    )
]
layout = go.Layout(
    title='Mt Bruno Elevation',
    autosize=False,
    width=500,
    height=500,
    margin=dict(
        l=65,
        r=50,
        b=65,
        t=90
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='elevations-3d-surface')
```

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(m, b):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    plt.plot(x, m * x**2 + b)
    plt.ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
