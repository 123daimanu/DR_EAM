```python
import numpy as num
import matplotlib.pyplot as plt
```

```python
def fEle_(r,data):
    re,fe,alpha,beta,A,B,kappa,lam=data
    return fe*num.exp(-beta*((r/re)-1))/(1+((r/re)-lam)**20)

def phiN_(r,data):
    re,fe,alpha,beta,A,B,kappa,lam=data
    return A*num.exp(-alpha*(r/re-1))/(1+((r/re)-kappa)**20)
    
                     
def phiE_(r,data):
    re,fe,alpha,beta,A,B,kappa,lam=data
    return -B*num.exp(-beta*((r/re)-1))/(1+((r/re)-lam)**20)
    

def phiT_(r,data):
    return phiN_(r,data) + phiE_(r,data)

def JM(r,data1,data2):
    fe1=fEle_(r,data1)
    phi1=phiT_(r,data1)
    
    fe2=fEle_(r,data1)
    phi2=phiT_(r,data1)
    
    jphi = 0.5* (fe1/fe2*phi2 + fe2/fe1 * phi1)
    return jphi
```

```python
#data = re, fe, alpha, beta, A, B, kappa, lambda

oxygen = 3.64857, 1.39478, 5.44072, 3.59746, 0.34900, 0.57438, 0.08007, 0.9310
OAl = 2.98520,0, 8.49741, 4.52114, 0.09738, 0.38121,0.18967, 0.95234
Al = 2.86392, 1.20378, 6.61317, 3.52702, 0.31487, 0.36555, 0.37985, 0.75969

r=num.linspace(0.1,5,1000)
```

```python
OAlPhi = phiT_(r,OAl)
OAlJM=JM(r,oxygen,Al)
```

```python
fig=plt.figure(figsize=(10,10))
plt.plot(r,OAlPhi)
plt.plot(r,OAlJM)
plt.xlim([0.5,3])
plt.ylim([-5,20])
```
