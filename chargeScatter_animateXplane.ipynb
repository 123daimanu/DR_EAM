```python
import numpy as num
import matplotlib.pyplot as plt

# Determine the index of the atoms having same z-value 
#list the index of atoms in same plane
fileOmd=open("Pt_test_7_layers_newc.xyz")
linesOmd=fileOmd.readlines()

posx=[]
for line in linesOmd:
    linesSplit=str.split(line)
    length=len(linesSplit)
    
    if(linesSplit[0]!="Na+"):
        posx.append(float(linesSplit[1]))
    
    else:
        break;
       
        


lengthPosx=len(posx)
sortPos=num.argsort(posx) # sorts the list with arguments as output i.e position as output



arrSortPos=num.array(sortPos)

# Make the different plane using the index of the atoms
indexPosxSort=[]
indexPosxSort.append(0)
init=posx[sortPos[0]]
for i in range(0,lengthPosx):
    if(posx[sortPos[i]]!=init):
        indexPosxSort.append(i)
    init=posx[sortPos[i]]


# X Y and C arrays for different atoms.

filexyz=open("Pt_test_7_layers_newc.xyz")
linesXYZ=filexyz.readlines()
atomNumber=lengthPosx;
layerCrystal=len(indexPosxSort)
numFrames=100
z=num.zeros((101,atomNumber))
y=num.zeros((101,atomNumber))
c=num.zeros((101,atomNumber))

count=0
index=0

for line in linesXYZ:
    lineApp=str.split(line);
    
    if(lineApp[0]=="Na+"):
        count=count+1
        index=0
        continue
    else:
        y[count][index]=float(lineApp[2])
        z[count][index]=float(lineApp[3])
        c[count][index]=float(lineApp[4])
        index=index+1
        
        
indexPosxSort.append(atomNumber)

#scatter plot of top layer
frame=100
Layer=layerCrystal
print indexPosxSort
for frameNo in range(frame+1):
    fig=plt.figure()
    for xLayer in range(1,Layer+1):
        layer=range(indexPosxSort[xLayer-1],indexPosxSort[xLayer])
        position=arrSortPos[layer]
        charge=c[frameNo,position]
        zz= z[frameNo,position]
        yy= y[frameNo,position]
        #scaled_c = (charge - charge.min()) / charge.ptp()
        #colors = plt.cm.coolwarm(scaled_c)
        sign=[];
        for var in charge:
            if (var<0):
                sign.append("blue");
            elif(var>0):
                sign.append("red");
            else:
                sign.append("black")
        ax1 = fig.add_subplot(2,Layer/2,xLayer)
        plt.scatter(zz,yy , c=charge, s=50, cmap='jet',edgecolor=sign,linewidth=1,vmin=-0.05,vmax=0.05)
        if (xLayer==Layer/2):
            plt. colorbar()
        plt.grid()
        
    fileName=str(frameNo)+".jpeg"
    plt.title(str(frameNo))
    plt.savefig(fileName)
       

```
