```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x =data1[:,0]
y =data1[:,1]
z =data1[:,2]


ax.scatter(x, y, z, c=z, marker='o',cmap="Reds")

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
#ax.view_init(50, 120)
plt.show()
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import numpy as np

x = data1[:,0]
y= data1[:,1]
z=data1[:,2]
trace1 = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=3,
        color=z,                # set color to an array/list of desired values
        colorscale='Reds',   # choose a colorscale
        opacity=1
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='3d-scatter-colorscale')
```

```python
file=open("/home/hbhattar/afs/Hemanta/metals/L10/minStrucutre.info",'w')
file.write("#name\tc\ta\tenergy\n")

names=["AgTi","AlMg","AlTi","AuCu","NiFe","NiMg","PdFe","PdTi","PtCo","PtFe","PtNi","ZrMg"]
for name in names:
    try:
        data1=num.loadtxt("../L10/%s.landscape"%(name))
        d=num.argsort(data1[:,2])
        plt.plot(data1[d,2],'o')
        plt.show()
        optStrucutre=data1[d][0]
        print(name,optStrucutre)
        file.write("%s\t%f\t%f\t%f\n"%(name,optStrucutre[0],optStrucutre[1],optStrucutre[2]))
    except:
        print("%s not found"%(name))
    
file.close()
```
