```python
import numpy as num
import matplotlib.pyplot as plt

# Determine the index of the atoms having same z-value 
#list the index of atoms in same plane
fileOmd=open("Pt_test_7_layers_newc.xyz")
linesOmd=fileOmd.readlines()

posy=[]
for line in linesOmd:
    linesSplit=str.split(line)
    length=len(linesSplit)
    
    if(linesSplit[0]!="Na+"):
        posy.append(float(linesSplit[1]))
    
    else:
        break;
       
        


lengthPosy=len(posy)
sortPos=num.argsort(posy) # sorts the list with arguments as output i.e position as output



arrSortPos=num.array(sortPos)

# Make the different plane using the index of the atoms
indexPosySort=[]
indexPosySort.append(0)
init=posy[sortPos[0]]
for i in range(0,lengthPosy):
    if(posy[sortPos[i]]!=init):
        indexPosySort.append(i)
    init=posy[sortPos[i]]


# X Y and C arrays for different atoms.

filexyz=open("Pt_test_7_layers_newc.xyz")
linesXYZ=filexyz.readlines()
atomNumber=lengthPosy;
layerCrystal=len(indexPosySort)
numFrames=100
z=num.zeros((101,atomNumber))
x=num.zeros((101,atomNumber))
c=num.zeros((101,atomNumber))

count=0
index=0

for line in linesXYZ:
    lineApp=str.split(line);
    
    if(lineApp[0]=="Na+"):
        count=count+1
        index=0
        continue
    else:
        x[count][index]=float(lineApp[2])
        z[count][index]=float(lineApp[3])
        c[count][index]=float(lineApp[4])
        index=index+1
        
        
indexPosySort.append(atomNumber)

#scatter plot of top layer
frame=100
Layer=layerCrystal
print indexPosySort
for frameNo in range(frame+1):
    fig=plt.figure()
    for yLayer in range(1,Layer+1):
        layer=range(indexPosySort[yLayer-1],indexPosySort[yLayer])
        position=arrSortPos[layer]
        charge=c[frameNo,position]
        zz= z[frameNo,position]
        xx= x[frameNo,position]
        scaled_c = (charge - charge.min()) / charge.ptp()
        colors = plt.cm.coolwarm(scaled_c)
        sign=[];
        for var in charge:
            if (var<0):
                sign.append("blue");
            elif(var>0):
                sign.append("red");
            else:
                sign.append("green")
        ax1 = fig.add_subplot(2,Layer/2,yLayer)
        plt.scatter(zz,xx , c=colors, s=50, cmap='jet_r',edgecolor=sign,linewidth=1)
        plt.grid()
        
    fileName=str(frameNo)+".jpeg"
    plt.title(str(frameNo))
    plt.savefig(fileName)
       

```
