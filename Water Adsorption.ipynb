```python

def V_mie(r,em,sm):
    return(4.5531*em*((sm/r)**11-(sm/r)**6))
def V_repulsive(r,er,sr):
    return er*(sr/r)**3
def plotting(r,em,sm,er,sr):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    plt.plot(r,V_mie(r,em,sm),label="Pt-O")
    plt.plot(r,V_repulsive(r,er,sr),label = "Pt-H")
    plt.plot(r,V_mie(r,em,sm)+V_repulsive(r,er,sr),label = "Pt-O + Pt-H")
    plt.legend(loc = 4)
    plt.ylim([-1,1])
    plt.grid()
    plt.show()
    
```

```python
import numpy as np
x=np.linspace(1,5,100)

plotting(x,.3,2.4,1,1)
```

```python
class PtOH():
    def __init__(self, em = 0, sm = 0, er = 0, sr = 0, n = 11, m = 6, r = 3):
        self.em = em
        self.sm = sm
        self.er = er
        self.sr = sr
        self.n = n
        self.m = m
        self.r = r
        self.xmin = 0.5
        self.xmax = 4
        self.ymin = -1
        self.ymax = 1
    def V_mie(self):
        return(4.5531 * self.em *(( self.sm / self.rr )**self.n - ( self.sm / self.rr )**self.m))
    def V_repulsive(self):
        return self.er *( self.sr / self.rr )**self.r
    def set_xlim(self,xmin,xmax):
        self.xmin = xmin
        self.xmax = xmax
    def set_ylim(self,ymin, ymax):
        self.ymin = ymin
        self.ymax = ymax
    
    def plotting(self):
        self.rr = np.linspace(self.xmin, self.xmax, 100)
        fig = plt.figure()
        ax = fig.add_axes([0,0,2,2])
        plt.plot(self.rr, self.V_mie(), label = "Pt-O")
        plt.plot(self.rr, self.V_repulsive(), label = "Pt-H")
        plt.plot(self.rr, self.V_mie() + self.V_repulsive(), linestyle = '--', label = "Pt-O + Pt-H")
        
        plt.ylim([self.ymin, self.ymax])
        plt.vlines(2.10,self.ymin,self.ymax,linestyle = 'solid', linewidth = 1, label = 'Atop')
        plt.vlines(2.89,self.ymin,self.ymax,linestyle = '--', linewidth = 1.3, label = 'Bridge')
        plt.vlines(3.02,self.ymin,self.ymax,linestyle = '-.', linewidth = 1.3, label = 'Hollow')
        plt.vlines(3.6,self.ymin,self.ymax,linestyle = '-', linewidth = 1.3, label = 'Repulsive Start')
        plt.legend(loc = 3)
        plt.xlabel("r")
        plt.ylabel("V")
        plt.grid()
        plt.show()
```

```python
import numpy as np
water_in_pt = PtOH(em = 5.5, sm = 3, er = 1, sr = 3, n = 15, m = 6, r = 6)
water_in_pt.set_ylim(-10,10)
water_in_pt.set_xlim(0.5,5)
water_in_pt.plotting()
```
