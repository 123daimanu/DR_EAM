```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "/home/hbhattar/afs/Hemanta/metals/ThermalConductivity/EqulibratedSystem/"
layers = list(range(1,19))
sides = ['L',"R"]
extension = ['.rcorrz']

system = ["MNWN","MFWN", "MNWN_TIP4P", "MFWN_TIP4P", "MNWF", "MFWF"]

systemListR = [path+sys+"_Dynamics_TimeAdjusted_"+str(l)+"_"+"R"+ex for sys in system for l in layers \
             for ex in extension]

systemListL = [path+sys+"_Dynamics_TimeAdjusted_"+str(l)+"_"+"L"+ex for sys in system for l in layers \
             for ex in extension]


systemList = [sys+"."+str(l)+ex for sys in system for l in layers \
             for ex in extension]
```

```python
def getData(file,threshold=25000*4):
    df = pd.read_csv(file,skiprows=7, delimiter="\t", names=["time","Corr"])
    df = df[df.time<threshold]
    return df


def getExpoFit(dfR,dfL,destName,savefigName="plot",savefig=False):
    from scipy.optimize import curve_fit
    import numpy as np
    
    
    def linear(x,a,b):
        return a * x + b
    
    
    
    coeff, varr =curve_fit(linear,list(dfR.time),list(dfR.Corr))
    
    aR,bR = tuple(coeff)
    
    coeff, varr =curve_fit(linear,list(dfL.time),list(dfL.Corr))
    
    aL,bL = tuple(coeff)
    
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()
    fig.set_size_inches(8,6)
    ax.set_title(destName)
    plt.plot(dfR.time,linear(dfR.time,aR,bR),color='red',label='R-fit')
    plt.plot(dfR.time, dfR.Corr,color='blue',label='R-data')
    
    plt.plot(dfR.time,linear(dfL.time,aL,bL),color='green',label='L-fit')
    plt.plot(dfL.time, dfL.Corr,color="black",label='L-data')
    plt.legend()
    if savefig:
        plt.savefig(destName+".pdf")
        
    return (aL+bL)/2
    
def getTitle(file):
    title =  file.split("/")[-1].replace(".","_")
    return title


def getPlots(dfList,nameList,savefig=False):
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()
    fig.set_size_inches(10,10)
    for index,df in enumerate(dfList):
        plt.plot(df.time, df.Corr,label=nameList[index], linewidth=2)
        
    plt.legend()    
    if savefig:
        plt.savefig(savefigName+"pdf")
    plt.show()
    
        
```

```python
titleList =  list(map(getTitle,systemList))



dfListR=list(map(getData,systemListR))
dfListL=list(map(getData,systemListL))


relax_time =  list(map(getExpoFit,dfListR, dfListL, titleList))
```

```python
file_Name
```

```python
relaxList=[]
for name in file_Name:
    values = []
    f = open(path+name+".dat",'w')
    system = name.split("-")[0]
    ext = name.split("-")[1]
    keys = [system+"."+str(num)+"."+ext for num in range(1,19)]
    for index,key in enumerate(keys):
        f.write("%d\t%f\n"%(index+1,relaxTime_dict[key]))
        values.append(relaxTime_dict[key])
    relaxList.append(values)
    f.close()
```

```python
relaxDict = dict(zip(file_Name, relaxList))
```
