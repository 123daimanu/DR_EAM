```python
import numpy as num
from sys import argv
IFile="../FDEAMCrystalCalculations/PtNew.xyz"
OFile="../FDEAMCrystalCalculations/PtCoord.dat"

# Program that edits .xyz file making the charge of sodium 0 so that we can see correct distribution of charge in the jmol.



xyzFile=open(IFile) #opens the initial .xyz file
outFile = open(OFile,'w')
coord=[]
lines=xyzFile.readlines()
linesAppend=[]
count = 0
for line in lines:
    lineApp=str.split(line);
    outFile.write("%s\t%s\t%s\n"%(lineApp[1],lineApp[2],lineApp[3]))
    coord.append([float(lineApp[1]),float(lineApp[2]),float(lineApp[3])])
    count =  count +1
outFile.close()
xyzFile.close()
```
