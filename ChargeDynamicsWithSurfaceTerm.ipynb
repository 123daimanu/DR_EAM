```python
import numpy as num
import pylab as lab
```

```python

pos=num.arange(100)
```

```python
cm=sum(pos)/len(pos)
```

```python
pos=pos-cm
length=abs(pos[0]-pos[-1])
beta=2*num.pi/length**3
```

```python
charge=num.random.rand(9)
```

```python
def dipole(pos,charge):
    return sum(pos*charge)
```

```python
def ESI(charge,pos,index):
    distance=abs(pos-pos[index])
    distance=distance[pos!=pos[index]]
    charge=charge[pos!=pos[index]]
    return sum(charge*distance)
```

```python
chargeList=[]
velocityList=[]
chPrev=num.zeros(len(pos))
velocityPrev=num.zeros(len(pos))
print(chPrev)
numberIteration=1
count=0
curv=10
E=1
M=10
chargeList.append(chPrev)
dt=0.0000001
while(count<=numberIteration):
    for var in range(len(pos)):
        
        accln=-((curv*chPrev[var]+2*beta*dipole(pos,num.array(chPrev))*pos[var]-E*pos[var])-ESI(num.array(chPrev),pos,var))/M
        print("index:%d,iter:%d,accln:%f"%(var,count,accln))
        vel=velocityPrev[var]+accln*dt
        print(vel)
        charge=(vel**2-velocityPrev[var]**2)/(2*accln)+chPrev[var]
        chPrev[var]=charge
        velocityPrev[var]=vel
    chargeList.append(chPrev)
    velocityList.append(velocityPrev)
    count=count+1
        
        
        
        
```

```python
lab.plot(pos,chargeList[-1],'o')
lab.show()
```
