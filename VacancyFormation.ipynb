```python
#==================================================================================================================
path="/home/hbhattar/afs/Hemanta/metals/VacancyFormation/SimulationFileTest.omd";                  #the path of the file
fileOutRelaxedRemoved=open("../VacancyFormation/SimulationFileTestRelaxRemoved.omd",'w')
fileOutRelaxed=open("../VacancyFormation/SimulationFileTestRelaxed.omd",'w')

#====================================================================================================================

natoms=108
nRevStr=str(natoms-1)
fileOmd=open(path)  #dump file for info extraction
lines=fileOmd.readlines()
```

```python
lineNewFile=[]
for line in lines:
    words=line.split()
        
    if nRevStr in words:
        continue
        
    if "useMinimizer" in words:
        words[2]="True"
        words.append("\n")
        line=" ".join(words)
        
    if "ensemble" in words:
        words[0]="//ensemble"
        words.append('\n')
        line=" ".join(words)
    
    lineNewFile.append(line)
fileOutRelaxedRemoved.writelines(lineNewFile)
fileOutRelaxedRemoved.close()   
```
